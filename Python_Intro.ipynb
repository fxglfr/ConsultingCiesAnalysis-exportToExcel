```python
import yfinance as yf
import pandas as pd

def get_stock_data(tickers=["TSLA"], period="1y"):
    data = yf.download(tickers, period=period)['Close']
    df = pd.DataFrame(data)
    return df

# Exemple d'utilisation de la fonction
tickers = ["TSLA"]
df = get_stock_data(tickers, "1y")
df
```

```python
# Crée un graphique à partir des données contenues dans le DataFrame 'df'
# Création d'un graphique en ligne pour les cours de l'action Tesla
plt.figure(figsize=(12, 6))
plt.plot(df, label='Prix de Tesla')
plt.title('Cours de l\'action Tesla (Données Fictives)')
plt.xlabel('Date')
plt.ylabel('Close')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Trouve la valeur maximale dans chaque colonne du DataFrame 'df'
df.max()
```

```python
import yfinance as yf
import pandas as pd

# Récupérer la liste des tickers du S&P 500
url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
table = pd.read_html(url)
df = table[0]
tickers = df['Symbol'].tolist()
```

```python
# Télécharger les données historiques pour tous les tickers
data = get_stock_data(tickers)

data.describe()
```

```python
# Quels sont les cinq tickers boursiers ayant les cours moyens les plus élevés parmi ceux que vous avez analysés ?
# Calculer la moyenne pour chaque ticker
mean_prices = data.mean()

# Trier ces moyennes dans l'ordre décroissant
sorted_mean_prices = mean_prices.sort_values(ascending=False)

# Sélectionner les cinq premiers tickers
top_five_tickers = sorted_mean_prices.head(5)

print("Les cinq tickers ayant les cours moyens les plus élevés sont :")
print(top_five_tickers)
```

```python
# Calcule la moyenne mobile sur 7 jours (7-day MA) en utilisant la fonction rolling() sur la colonne 'Close' et en prenant la moyenne
# Calcul de la moyenne mobile sur 7 jours pour la colonne 'Close'
df['7_day_MA'] = df['Close'].rolling(window=7).mean()

df['7_day_MA'].head()
```
